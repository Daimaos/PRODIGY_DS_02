## üì• Step 1: Import Required Libraries


```python
# Data Manipulation
import pandas as pd
import numpy as np

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
```

## üìë Step 2: Load the Titanic Dataset


```python
# Load dataset (assumes CSV file is available locally as 'titanic.csv')
df = pd.read_csv('titanic.csv')

# Display first 5 records
df.head()
```

## üìù Step 3: Explore Dataset Structure and Info


```python
# Dataset structure
df.info()
```

```python
# Check summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

## üßπ Step 4: Data Cleaning


```python
# Fill missing 'Age' values with median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing 'Embarked' values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' due to high missing data
df.drop(columns='Cabin', inplace=True)

# Confirm no missing values remain
df.isnull().sum()
```

## üìä Step 5: Exploratory Data Analysis (EDA)


```python
# Bar Plot: Survival Count
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar Plot: Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Histogram: Age Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot: Age by Pclass
sns.boxplot(x='Pclass', y='Age', data=df, palette='Pastel1')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Heatmap: Correlation Matrix
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Heatmap: Correlation Matrix (only numeric columns)
plt.figure(figsize=(10,6))

# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Generate heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```

## üìä Additional EDA Visualizations


```python
# Countplot: Gender Distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', data=df, palette='pastel')
plt.title('Passenger Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Countplot: Survival by Gender
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(labels=['Did Not Survive', 'Survived'])
plt.show()
```

```python
# Histogram: Age Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Countplot: Survival by Pclass
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(labels=['Did Not Survive', 'Survived'])
plt.show()
```

```python
# Violinplot: Fare vs Pclass
plt.figure(figsize=(10,6))
sns.violinplot(x='Pclass', y='Fare', data=df, palette='muted')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
# Pie Chart: Survival Rate
plt.figure(figsize=(6,6))
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], labels=['Did Not Survive','Survived'])
plt.title('Overall Survival Rate')
plt.ylabel('')
plt.show()
```

## üìå Overall Conclusion:

- Conducted comprehensive exploratory data analysis (EDA) on the Titanic dataset to understand passenger demographics and survival patterns.
- Identified key relationships between variables such as **Gender**, **Passenger Class (Pclass)**, **Age**, and **Fare** with survival outcomes.
- Visualizations revealed that:
  - Females had a significantly higher survival rate compared to males.
